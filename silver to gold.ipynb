```python
path='/mnt/silver/silver/'
df = spark.read.format('delta').load(path)
```

```python
display(df)
```

```python
#Drop những cột không cần thiết
cols_to_drop = [ 'number_of_persons_injured', 'number_of_pedestrians_killed',
                'number_of_cyclist_injured', 'number_of_motorist_injured', 'number_of_motorist_killed', 'number_of_persons_killed', 'number_of_pedestrians_injured',
                'number_of_cyclist_killed','zip_code','on_street_name','off_street_name','cross_street_name','crash_date','crash_time']
df= df.drop(*cols_to_drop)
```

```python
#Ma hoa du lieu bien phan loai
from pyspark.sql import DataFrame
from pyspark.ml.feature import StringIndexer
categorical_cols = ['contributing_factor_vehicle_1', 'contributing_factor_vehicle_2','contributing_factor_vehicle_3','contributing_factor_vehicle_4','contributing_factor_vehicle_5',
                    'vehicle_type_code1', 'vehicle_type_code2', 'borough']

def encode_categorical_cols(df: DataFrame, categorical_cols: list) -> DataFrame:
    for col_name in categorical_cols:
        indexer = StringIndexer(inputCol=col_name, outputCol=col_name + "_encoded", handleInvalid='keep')
        df = indexer.fit(df).transform(df)
    return df

df_encoded = encode_categorical_cols(df, categorical_cols)
```

```python
df_final = df_encoded.drop(*categorical_cols)
display(df_final)
```

```python
df_final.describe().toPandas().transpose()
```

```python
#Lưu xuống tầng gold
output_path= '/mnt/gold/gold/'
df_final.write.format('delta').mode('overwrite').save(output_path)
```
